# Shuffle/Batch/Repeat

```python
data = tf.data.Dataset.range(10)
# uncomment any of the below (also multiple ones if you want) and run this cell each time. 
# make sure you understand what each transformation does!

data = data.shuffle(10)  # 10 is the buffer size -- play with this!!

elems = 0
for thing in data:
    elems += 1
    print(thing.numpy())
print("\nTotal number of elements: {}".format(elems))

data = data.batch(2)  # 2 is batch size -- change it!


elems = 0
for thing in data:
    elems += 1
    print(thing.numpy())
print("\nTotal number of elements: {}".format(elems))

data = data.repeat(2)  # number of repetitions


elems = 0
for thing in data:
    elems += 1
    print(thing.numpy())
print("\nTotal number of elements: {}".format(elems))
```

-shuffles numbers 1 to 10
-puts numbers in batches
-loads the dataset again + shuffles again and batches it again
